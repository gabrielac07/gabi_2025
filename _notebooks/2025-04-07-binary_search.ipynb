## PART A: Binary Breakdown

### Convert these decimals to binary:

### 1. **42**

#### Division by 2 steps:
```
42 ÷ 2 = 21 remainder 0  
21 ÷ 2 = 10 remainder 1  
10 ÷ 2 = 5  remainder 0  
5 ÷ 2  = 2  remainder 1  
2 ÷ 2  = 1  remainder 0  
1 ÷ 2  = 0  remainder 1  
→ Binary: 101010
```

| 2⁵ | 2⁴ | 2³ | 2² | 2¹ | 2⁰ |
|-----|-----|-----|-----|-----|-----|
| 1   | 0   | 1   | 0   | 1   | 0   |

## Free Response Questions

---

### Why is binary search more efficient for large data than linear search?
Binary search cuts the list in half with each step, reducing the number of comparisons needed. In contrast, linear search checks each item one-by-one, which takes longer as the list grows.

---

### What happens if the list isn’t sorted and you try to use binary search?
If the list isn’t sorted, binary search won’t work correctly. It might skip over the target or give the wrong result because it relies on the order of elements.

---

### Could you use binary search in a video game leaderboard or streaming service search bar? Why or why not?
Yes, if the leaderboard or search suggestions are **sorted**, binary search can make searching much faster. It improves performance and gives results quickly even in huge databases.
